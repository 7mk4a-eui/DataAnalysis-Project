```python
# Google Colab Notebook for Survey Data Analysis

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Upload the dataset
from google.colab import files
uploaded = files.upload()

# Load the dataset
file_name = list(uploaded.keys())[0]  # Get the uploaded file name
df = pd.read_csv(file_name)

# Display first few rows to understand the data structure
df.head()

# Clean column names (strip spaces)
df.columns = df.columns.str.strip()

# Mapping GPA and Study Hours to numeric values
gpa_mapping = {
    "Below 2": 1.75,
    "2.0 - 2.5": 2.25,
    "2.5 - 3.0": 2.75,
    "3.0 - 3.5": 3.25,
    "3.5 - 4.0": 3.75
}

df["GPA"] = df["What is your current cumulative GPA?"].map(gpa_mapping)

study_hours_mapping = {
    "Less than 1 hour": 0.5,
    "1-2 hours": 1.5,
    "3-4 hours": 3.5,
    "5+ hours": 5
}

df["Study Hours"] = df["How many hours do you study outside of class per day?"].map(study_hours_mapping)

# Scatter plot: Study Hours vs. GPA
plt.figure(figsize=(8,5))
sns.scatterplot(x=df["Study Hours"], y=df["GPA"])
plt.xlabel("Study Hours per Day")
plt.ylabel("GPA")
plt.title("Study Hours vs. GPA")
plt.grid(True)
plt.show()

# Calculate correlation between Study Hours and GPA
correlation = df["Study Hours"].corr(df["GPA"])
print(f"Correlation between Study Hours and GPA: {correlation:.2f}")

# Bar chart: Study Frequency Before Exams
plt.figure(figsize=(8,5))
sns.countplot(y=df["How often do you study the material before the exam period?"], order=df["How often do you study the material before the exam period?"].value_counts().index)
plt.xlabel("Count")
plt.ylabel("Study Frequency Before Exams")
plt.title("Distribution of Study Frequency Before Exams")
plt.show()

# Bar chart: GPA distribution
plt.figure(figsize=(8,5))
sns.countplot(y=df["What is your current cumulative GPA?"], order=df["What is your current cumulative GPA?"].value_counts().index)
plt.xlabel("Count")
plt.ylabel("GPA Categories")
plt.title("Distribution of GPA Among Students")
plt.show()

# Summary statistics
print("\nSummary Statistics for Study Hours:")
print(df["Study Hours"].describe())

print("\nSummary Statistics for GPA:")
print(df["GPA"].describe())
```
